# Binary Classification with Keras

### Classify loan Approval
1. 0 - Rejected
2. 1 - Approved

```python
url1 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/train_loan.csv"
print(url1)
```

```python
import pandas as pd
df = pd.read_csv(url1)
df.head()
```

## Step 2 - Perform Basic Data Quality Checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

## Step 3 - Seperate X and Y(Loan_Status)

```python
X = df.drop(columns= ["id", "loan_status"])
Y = df[["loan_status"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind="bar")
```

## Step 4 - Apply preprocessing on X

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False)
)
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe, con),
        ("cat", cat_pipe, cat)
    ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

## Step 5 - Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.shape[1]
```

## Step 6 - Build the Neural Network

image.png

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
```

```python
# Initialize neural network
model = Sequential()

# Add Input layer
model.add(Input( shape=(xtrain.shape[1],) ))

# Add Dense layer1
model.add(Dense(units=32, activation="relu"))
#Add Dropout layer1
model.add(Dropout(rate=0.3))

# Add Dense layer2
model.add(Dense(units=16, activation="relu"))
#Add Dropout layer2
model.add(Dropout(rate=0.2))

# Add output layer with sigmoid activation
model.add(Dense(units=1, activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=60)
```

## Plot the learning Curve

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.plot(hist.history["loss"], label="Train Loss")
plt.plot(hist.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Binary cross Entropy loss")
plt.title("Learning Curve")
plt.legend()
plt.show()
```

## Step 7 - Model Evaluation

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

```python
# Predict Probability for train and test
yprob_train = model.predict(xtrain)
yprob_test = model.predict(xtest)
```

```python
ypred_train = [1 if prob >= 0.5 else 0 for prob in yprob_train]
ypred_train[0:5]
```

```python
# Create a function to give final prediction
def predict_results(model, x):
  yprob = model.predict(x)
  ypred = [1 if prob >= 0.5 else 0 for prob in yprob]
  return ypred, yprob
```

```python
ypred_train, yprob_train = predict_results(model, xtrain)
```

```python
ypred_train[0:5]
```

```python
ypred_test, yprob_test = predict_results(model, xtest)
```
